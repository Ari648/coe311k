```python
import math
import random
import numpy
import matplotlib.pyplot as plt

def montecarlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range(0,number_of_shots):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
        if (x*x + y*y <1):
            number_of_hits = number_of_hits + 1
        
    return (4*number_of_hits/number_of_shots)
```

```python
trueError = []
relativeError = []

for i in range(0, 100):
    trueError.append(abs(numpy.pi - montecarlo_pi(1000)))
    relativeError.append(abs((numpy.pi - montecarlo_pi(1000))/numpy.pi))
    

plt.plot(trueError, 'g', relativeError, 'r')
plt.show()
```
